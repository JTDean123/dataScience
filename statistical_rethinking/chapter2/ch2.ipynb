```python
def water_posterior(observations, 
                    prob_grid = np.arange(0, 1.001, 0.01),
                    prior=1):
    
    binary_obs = [1 if i == 'W' else 0 for i in observations]
    number_of_observations = len(binary_obs)
    number_of_successes = sum(binary_obs)
    
    likelihoods = []
    for prob in prob_grid:
        likelihood = scipy.stats.binom.pmf(number_of_successes,
                                           number_of_observations,
                                           prob)
        
        likelihoods.append(likelihood*prior)
    
    posterior_distribution = [i/sum(likelihoods) for i in likelihoods]
    
    plt.figure(figsize=(6,4))
    plt.title(observations)
    plt.plot(prob_grid, posterior_distribution)
    plt.grid()
    plt.show()
```

```python
# 1, 2, and 3
water_posterior('WWW')
water_posterior('WWWL')
water_posterior('LWWLWWW')
```

```python
# assume prior is zero when p < 0.5 and postive constant otherwise
def water_posterior(observations, 
                    prob_grid = np.arange(0, 1.001, 0.01),
                    prior=1):
    
    binary_obs = [1 if i == 'W' else 0 for i in observations]
    number_of_observations = len(binary_obs)
    number_of_successes = sum(binary_obs)
    
    likelihoods = []
    for prob in prob_grid:
        likelihood = scipy.stats.binom.pmf(number_of_successes,
                                           number_of_observations,
                                           prob)
        
        if prob < 0.5: 
            prior = 0
        else:
            prior = 1
            
        likelihoods.append(likelihood*prior)
    
    posterior_distribution = [i/sum(likelihoods) for i in likelihoods]
    
    plt.figure(figsize=(6,4))
    plt.title(observations)
    plt.plot(prob_grid, posterior_distribution)
    plt.grid()
    plt.show()


```

```python
water_posterior('WWW')
water_posterior('WWWL')
water_posterior('LWWLWWW')
```
