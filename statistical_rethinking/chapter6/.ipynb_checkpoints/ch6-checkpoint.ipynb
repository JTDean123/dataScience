```python
data = pd.read_csv('../chapter4/howell.txt', sep=';')

# normalize the age
data['age'] = (data['age'] - data['age'].mean())/data['age'].std()

data.head(2)
```

```python
# split the data in half
train, test = train_test_split(data, test_size=0.5, random_state=123)
```

```python
plt.scatter(train['age'], train['height'], alpha=0.3)
plt.xlabel('age')
plt.ylabel('height')
plt.grid()
```

```python
# fit a bunch of models to predict height as a function of age
```

```python
train['height'].std()
```

```python
# model 1
with pm.Model() as model_1:

    # priors
    sigma = pm.Uniform('sigma', lower=0, upper=30)
    alpha = pm.Normal('alpha',  mu=150, sigma=20)
    b1 = pm.Normal('b1', mu=0, sigma=10)
    
    # deterministic
    mu = alpha + b1 * train['age'].values
    
    # likelihood
    height = pm.Normal('height', mu, sigma, observed=train['height'])
    
    trace_1 = pm.sample(1000)
    waic_1 = pm.waic(trace_1, model_1).waic
```

```python
# model 2
with pm.Model() as model_2:

    # priors
    sigma = pm.Uniform('sigma', lower=0, upper=30)
    alpha = pm.Normal('alpha',  mu=150, sigma=20)
    b1 = pm.Normal('b1', 0, 10)
    b2 = pm.Normal('b2', 0, 10)
    
    # deterministic
    mu = alpha + b1*train['age'] +  b2*train['age']**2
    
    # likelihood
    height = pm.Normal('height', mu, sigma, observed=train['height'])
    
    trace_2 = pm.sample(1000)
    waic_2 = pm.waic(trace_2, model_2).waic
```

```python
# model 3
with pm.Model() as model_3:

    # priors
    sigma = pm.Uniform('sigma', lower=0, upper=30)
    alpha = pm.Normal('alpha',  mu=150, sigma=20)
    b1 = pm.Normal('b1', 0, 10)
    b2 = pm.Normal('b2', 0, 10)
    b3 = pm.Normal('b3', 0, 10)
    
    # deterministic
    mu = alpha + b1*train['age'] +  b2*train['age']**2 + b3*train['age']**3
    
    # likelihood
    height = pm.Normal('height', mu, sigma, observed=train['height'])
    
    trace_3 = pm.sample(1000)
    waic_3 = pm.waic(trace_3, model_3).waic
```

```python
# model 4
with pm.Model() as model_4:

    # priors
    sigma = pm.Uniform('sigma', lower=0, upper=30)
    alpha = pm.Normal('alpha',  mu=150, sigma=20)
    b1 = pm.Normal('b1', 0, 10)
    b2 = pm.Normal('b2', 0, 10)
    b3 = pm.Normal('b3', 0, 10)
    b4 = pm.Normal('b4', 0, 10)
    
    # deterministic
    mu = alpha + b1*train['age'] +  b2*train['age']**2 + b3*train['age']**3 + b4*train['age']**4
    
    # likelihood
    height = pm.Normal('height', mu, sigma, observed=train['height'])
    
    trace_4 = pm.sample(1000)
    waic_4 = pm.waic(trace_4, model_4).waic
```

```python
# model 5
with pm.Model() as model_5:

    # priors
    sigma = pm.Uniform('sigma', lower=0, upper=30)
    alpha = pm.Normal('alpha',  mu=150, sigma=20)
    b1 = pm.Normal('b1', 0, 10)
    b2 = pm.Normal('b2', 0, 10)
    b3 = pm.Normal('b3', 0, 10)
    b4 = pm.Normal('b4', 0, 10)
    b5 = pm.Normal('b5', 0, 10)
    
    # deterministic
    mu = alpha + b1*train['age'] +  b2*train['age']**2 + b3*train['age']**3 + b4*train['age']**4 + \
            b5*train['age']**5
    
    # likelihood
    height = pm.Normal('height', mu, sigma, observed=train['height'])
    
    trace_5 = pm.sample(1000)
    waic_5 = pm.waic(trace_5, model_5).waic
```

```python
# model 6
with pm.Model() as model_6:

    # priors
    sigma = pm.Uniform('sigma', lower=0, upper=30)
    alpha = pm.Normal('alpha',  mu=150, sigma=20)
    b1 = pm.Normal('b1', 0, 10)
    b2 = pm.Normal('b2', 0, 10)
    b3 = pm.Normal('b3', 0, 10)
    b4 = pm.Normal('b4', 0, 10)
    b5 = pm.Normal('b5', 0, 10)
    b6 = pm.Normal('b6', 0, 10)
    
    # deterministic
    mu = alpha + b1*train['age'] +  b2*train['age']**2 + b3*train['age']**3 + b4*train['age']**4 + \
            b5*train['age']**5 + b6*train['age']**6
    
    # likelihood
    height = pm.Normal('height', mu, sigma, observed=train['height'])
    
    trace_6 = pm.sample(1000)
    waic_6 = pm.waic(trace_6, model_6).waic
```

```python
# these WAIC values seem different than what is in the book.  calculate them from scratch....
# find info about logsumexp here
# https://mc-stan.org/docs/2_23/stan-users-guide/log-sum-of-exponentials.html
```

```python
# start with trace_1
# need to get the likelihood for the data for each sample from the posterior

# get mu_i, for each data point across all of the samples from the posterior
mu_1 = map(lambda x: trace_1['alpha'] + trace_1['b1']*x, train['age'])

ll_ppd = []; pWAIC = []
for obs, mu in zip(train['height'], mu_1):
    ll_ = scipy.stats.norm.logpdf(obs, mu,  trace_1['sigma'])
    
    # log-pointwise predictive density
    ll_ppd_ = logsumexp(ll_) - np.log(ll_.shape[0])
    ll_ppd.append(ll_ppd_)
    
    # effective number of parameters
    pWAIC_ = np.var(ll_)
    pWAIC.append(pWAIC_)
    
# now combine
WAIC = -2*(sum(ll_ppd) - sum(pWAIC))
WAIC
```

```python
# compare to trace_4, the best model supposedly
mu_4 = map(lambda x: trace_4['alpha'] + trace_4['b1']*x + trace_4['b2']*x**2 + \
           + trace_4['b3']*x**3  + trace_4['b4']*x**4 , train['age'])

ll_ppd = []; pWAIC = []
for obs, mu in zip(train['height'], mu_4):
    ll_ = scipy.stats.norm.logpdf(obs, mu,  trace_4['sigma'])
    
    # log-pointwise predictive density
    ll_ppd_ = logsumexp(ll_) - np.log(ll_.shape[0])
    ll_ppd.append(ll_ppd_)
    
    # effective number of parameters
    pWAIC_ = np.var(ll_)
    pWAIC.append(pWAIC_)
    
# now combine
WAIC = -2*(sum(ll_ppd) - sum(pWAIC))
WAIC
```

```python
# model averaged predictions on train data
waic_weights = waic_df.reset_index().rename(columns={'index':'model'})
# sort to make life easier
waic_weights = waic_weights.sort_values(by=['model']).reset_index(drop=True)
waic_weights
```

```python
traces = [trace_1, trace_2, trace_3, trace_4, trace_5, trace_6]
models = [model_1, model_2, model_3, model_4, model_5, model_6]


ppc_w = pm.sample_posterior_predictive_w(traces, 2000, models,
                        weights=waic_weights['weight'].values,
                        progressbar=True)
```

```python
# we have 1000 samples for each train observation
# plot the mean and 97% interval for comparison
post_pred = pd.DataFrame({'age':train['age']})

# matrix is pred x observation
means = []; lowers = []; uppers = [] 
for obs in range(ppc_w['height'].shape[1]):
    
    pred = ppc_w['height'][:, obs]
    mean = pred.mean()
    lower_, upper_ = pm.stats.hpd(pred, alpha=0.97)
    means.append(mean)
    lowers.append(lower_)
    uppers.append(upper_)
    
    
post_pred['mean'] = means
post_pred['lower'] = lowers
post_pred['upper'] = uppers
post_pred.head()
```
